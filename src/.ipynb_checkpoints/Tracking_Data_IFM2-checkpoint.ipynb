```python
## IMPORT FUNCTIONS

import numpy as np # Package fro the array structures 
import pandas as pd
np.set_printoptions(suppress=True) # Supress scientific notation when printing
import matplotlib.pyplot as plt # This is the package for all the basic plotting (plots, histograms, scatters...)
from scipy.ndimage import gaussian_filter1d

import networkx as nx # Package for graph represenations 

import re 

from networkx.drawing.nx_agraph import graphviz_layout, to_agraph, write_dot
import pygraphviz as gv

import datetime
```

```python
## Set file names and paths

# I usually have a directory for each project, with subfolders at least for
# for data (data), scripts (src), plots and other results (results), hence the paths specified below

# Raw data directory
path2data = '../data/'
# Results directory
path2results = '../results/'
# Raw data file
rat_filename = '20210628_Rat5.txt'
# Data import log file
today = datetime.date.today()
log_filename = 'import_log_' + today.strftime("%Y%m%d") + '.txt'
# Maze edgelists
edgelist_filename = '../data/graph_edgelist.dat'
```

```python
### load graph
print("load graph")
G = nx.read_edgelist(edgelist_filename)  
print("graph: number of nodes = ",G.number_of_nodes(),", edges = ",G.number_of_edges())

### pre-compute all possible shortest path lengths and save in dictionary (does not save shortest paths itself)
DD = nx.shortest_path_length(G) 
DD = dict(DD)
print("dictionary: len =", len(DD), ", total items =", sum([len(dv) for dv in DD.values()]))

# get node positions from recosntructed graph 
pos = graphviz_layout(G)
```

```python
# Read file using "with" to avoid needing to close file afterwards
with open(path2data + rat_filename, 'r') as f:
    # Load all lines from file
    list_all = [x for k,x in enumerate(f.readlines())]
        
# Create/Overwrite and start writing log file    
with open(path2results + log_filename, 'w') as log_f:
    log_f.write('Import date: ' + today.strftime("%d-%m-%y") + '\n')
    log_f.write(list_all[0])

# Open to append logs
log_f = open(path2results + log_filename, 'a') 

# Locate trial line boundaries and first trial line
list_tr_bb = [n for n, x in enumerate(list_all) if re.match(r'^Summary Trial', x)]
trial_headers = [x for n, x in enumerate(list_all) if re.match(r'^Summary Trial', x)]

log_f.write( str(len(list_tr_bb)) + ' trials found.\n' )

# Add a final entry to list_tr_bb to mark end boundary of last trial
list_tr_bb.append(len(list_all))
```

```python
### DO SOME DATA CLEANING (Remove duplicate trials and headers)

# Look for duplicate trials to write info to log
dupe_trials = [x for n, x in enumerate(trial_headers) if x in trial_headers[:n]] # Get duplicate trial headers
log_f.write('Found ' + str(len(dupe_trials)) + ' duplicate trials: \n')
[log_f.write(x) for x in dupe_trials]

# Look for duplicate lines in general and remove them from list_all
dupe_lines = [n for n, x in enumerate(list_all) if x in list_all[:n]]

# Remove duplicate lines from from list_all
log_f.write('Found ' + str(len(dupe_lines)) + ' duplicate lines.\n')
for x in sorted(dupe_lines, reverse=True):
    # log_f.write('line: ' + str(x) + '\n')
    del list_all[x]

# Update trial boundaries after deletions
list_tr_bb = [n for n, x in enumerate(list_all) if re.match(r'^Summary Trial', x)]

log_f.write('Number of trials left after deletions: ' + str(len(list_tr_bb)) + '\n')

# Add a final entry to list_tr_bb to mark end boundary of last trial
list_tr_bb.append(len(list_all))
    
```

```python
# Load step-wise trial information 
list_sess = []
for l,tr in enumerate(list_tr_bb[:-1]): 
    # Find all lines starting with ( between trial boundaries
    list_sing = [x for x in list_all[list_tr_bb[l]:list_tr_bb[l+1]] if x[0]=='('] # Lines starting with ( 
    # Remove parenthesis, commas and '' from string
    list_sing = [x.replace("(", "") for x in list_sing]
    list_sing = [x.replace(")", "") for x in list_sing]
    list_sing = [x.replace(",", "") for x in list_sing]
    list_sing = [x.replace("'", "") for x in list_sing]
    list_sing = [x.replace("\n", "") for x in list_sing]
    # Now separate data in each line within the trial
    list_sing = [x.split(sep=' ') for x in list_sing]
    list_sess.append(list_sing)
    
log_f.write('Number of trials loaded: ' + str(len(list_sess)) + '\n')

log_f.close()
```

```python
# Now collect data for selected trials
# I wasn't sure what kind of data structure would be more useful for storing data with trial structure, maybe
# a dictionary would have been good? I went for a numpy array for each data point, with a first column indicating
# trial no., and a second one for the value:

size = sum([len(x) for x in list_sess]) # array shape: total no. of lines (across all trials)
trial_s = np.zeros(size, dtype=int)
first_node = np.zeros(size, dtype=int)
second_node = np.zeros(size, dtype=int)
first_timestamp = np.zeros(size)
second_timestamp = np.zeros(size)
seconds = np.zeros(size) # seconds calculated from converted timestamp
seconds_ff = np.zeros(size) # seconds from file
edge_distance = np.zeros(size)
speed_ff = np.zeros(size)

size = sum([len(x) for x in list_sess])+len(list_sess) # array shape: total no. of lines (across all trials), 
                                                # with one additional line per trial (25 trials in total) to flatten structure
trial_l = np.zeros(size, dtype=int)
cum_distance = np.zeros(size) # cumulative distance travelled within each trial
cum_seconds = np.zeros(size) # cumulative time within each trial
nodes = np.zeros(size, dtype=int) # nodes the rat passes through (flat)
speed = np.zeros(size)

# For the timestamps we need a bigger array with 3 cols: hours, minutes, seconds
shape = (3, sum([len(x) for x in list_sess])+len(list_sess))
timestamps = np.zeros(shape) # all timestamps (flat)

#tr_list = np.arange(10,11) # Select Trials #### WHY only this trial subset??

# Get info from each line in list_sess and store in the recently created arrays
line_no = 0 # initialize counters
cum_line_no = 0

for tr, trial in enumerate(list_sess):

    # Add inital trial line for nodes and timestamps array with starting node (first value of first line in the trial)
    trial_l[cum_line_no] = tr+1
    nodes[cum_line_no] = list_sess[tr][0][0]
    time_list = (list_sess[tr][0][2]).split(':')
    timestamps[:, cum_line_no] = [float(x) for x in time_list]
    # Initialize distance and seconds
    # cum_distance[0, cum_line_no] = tr+1
    # cum_seconds[0, cum_line_no] = tr+1
    # speed[0, cum_line_no] = tr+1
    cum_line_no +=1
    
    for x in list_sess[tr]:
        trial_s[line_no] = tr+1
        trial_l[cum_line_no] = tr+1
        # Read nodes
        first_node[line_no] = x[0] # Value of first column (tr+1) corresponds to trial no., starting with 1 instead of 0
        second_node[line_no] = x[1] # x contains a row of data. Each position corresponds with one data type
        nodes[cum_line_no] = x[1]
        
        # Read time in first timestamp and convert to seconds
        time_list = (x[2]).split(':')
        time_list_fl = [float(x) for x in time_list] # Convert to float
        first_timestamp[line_no] = ((time_list_fl[0] * 3600) + (time_list_fl[1] * 60) + time_list_fl[2])
        # Same for second timestamp:
        time_list = (x[3]).split(':')
        time_list_fl = [float(x) for x in time_list]
        second_timestamp[line_no] = ((time_list_fl[0] * 3600) + (time_list_fl[1] * 60) + time_list_fl[2])
        seconds[line_no] = round((second_timestamp[line_no] - first_timestamp[line_no]), 2)
        timestamps[:, cum_line_no] = time_list_fl
        
        # Read rest of info
        seconds_ff[line_no] = x[4]
        edge_distance[line_no] = x[5]
        speed_ff[line_no] = x[6]
        speed[cum_line_no] = x[6]
        
        # Add trial number for rows in cum_distance and cum_time. 
        # Distance and time values to be calculated later
        # cum_distance[0, cum_line_no] = tr+1
        # cum_seconds[0, cum_line_no] = tr+1

        line_no += 1
        cum_line_no += 1
        
    # cumulative distance travelled in this trial
    tr_cum_distance = np.cumsum(edge_distance[trial_s == tr+1]) 
    # Add to cum_distance padded with a 0 (initial position)
    cum_distance[trial_l == tr+1] = np.pad(tr_cum_distance, [(1, 0)], mode='constant')
    #Same for seconds
    tr_cum_seconds = np.cumsum(seconds[trial_s == tr+1])
    cum_seconds[trial_l==tr+1] = np.pad(tr_cum_seconds, [(1,0)], mode='constant')

#nodes = nodes.astype(int)

```

```python
## Some sanity checks

# Last second_node in all trials must be 104
for tr, trial in enumerate(list_sess):
    tr_nodes = second_node[trial_s == tr+1]
    if tr_nodes[-1] != 104:
        print('ERROR in Trial ' + str(tr+1) + 'Last node is not 104') 

# Difference in timestamps in seconds must be equal to seconds in file
if np.any(seconds_ff != seconds): 
    print('ERROR in calculation of seconds')
# This is giving an error, but visual inspection of the output of seconds_ff[1,:] - seconds[1,:] looks ok
# I think it might be due to rounding off errors giving small differences between values

# Check times monotonically increase (Very high speeds problem)
# ... 


```

```python
# Speed plots look very noisy. Let's try some smoothing with moving average window

speed_sm = np.zeros(speed.shape)
line_no = 0
for tr in range(1,26,1):
    speeds_tr = speed[trial_l==tr]
    speed_sm[line_no] = speeds_tr[1]
    line_no += 1 #Smoothed speed for first step in each trial is equalled to first speed
    
    for i in range(1, len(speeds_tr)-1, 1):
        speed_sm[line_no] = np.mean(speeds_tr[i-1:i+2])
        line_no +=1
    speed_sm[line_no] = speeds_tr[-1] # Last element equal to point speed
    line_no += 1
     

arange = range(3,18,1) #It would probably be nicer to do it with some kind of subplot so they are displayed side by side
for tr in arange:
    x = range(0, len(speed_sm[trial_l==tr]), 1)
    y = speed_sm[trial_l==tr]
    plt.plot(x, y, label=str(tr));
    plt.xlabel('Step');
    plt.ylabel('Speed (m/s)');
    plt.legend();
    plt.pause(0.3)
```
