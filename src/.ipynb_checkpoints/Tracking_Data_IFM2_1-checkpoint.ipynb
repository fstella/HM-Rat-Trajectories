# Rat hexMaze Analysis
The objective of this script is to identify time intervals of interest in rat maze navigation for posterior analysis of neural data. 

We will look at speed profiles in each run, U-turns, and deviations or come-backs to optimal (shortest) paths.

The current script will look at Rat 5 on day 2021-06-28

```python
## Import functions

import numpy as np  
import pandas as pd
np.set_printoptions(suppress=True) # Supress scientific notation when printing
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 5]
import seaborn as sns
from scipy.ndimage import gaussian_filter1d
import re # Regular expressions
import networkx as nx # Package for graph represenations 
from datetime import datetime, time 
import pygraphviz as gv
```

```python
## Set file names and paths

# I usually have a directory for each project, with subfolders at least for
# for data (data), scripts (src), plots and other results (results), hence the paths specified below

# Raw data directory
path2data = '../data/'
# Results directory
path2results = '../results/'
# Raw data file
rat_filename = '20210628_Rat5.txt'
# Data import log file
today = datetime.now()
log_filename = 'import_log_' + today.strftime("%Y%m%d") + '.txt'
# Maze edgelists
edgelist_filename = '../data/graph_edgelist.dat'
```

### Build a maze representation

Networkx library allows working with node graphs. The read_edgelist method reconstructs automatically maze geometry with certain acuracy, taking as input file with node connections and distance between nodes.

There is also a method to compute shortest paht lengths.

```python
## BUILD A MAZE REPRESENTATION:
# Load hexMaze graph and compute shortest paths

print("load graph")
G = nx.read_edgelist(edgelist_filename)  
print("graph: number of nodes = ",G.number_of_nodes(),", edges = ",G.number_of_edges())

### pre-compute all possible shortest path lengths and save in dictionary (does not save shortest paths itself)
DD = nx.shortest_path_length(G) 
DD = dict(DD)
print("dictionary: len =", len(DD), ", total items =", sum([len(dv) for dv in DD.values()]))

```

### Read data file and do some data cleaning

Read data into list of strings corresponding to all file lines. Find trial boundaries (trial headers) and record positions. Do some data cleaning and verification. Import results written to logfile.

Data will be stored in a pandas dataframe, transforming the 'wide' format in the file,
that represents each trajectory (start node, end node, edge speed, edge distance) into 
'long' format, with all the successive nodes in one column and corresponding info. 

```python
## LOAD RAT DATA from file 
# Store as list of strings (file lines) and locate trial boundaries

# Read file using "with" to avoid needing to close file afterwards
# Store each line in list_all
with open(path2data + rat_filename, 'r') as f:
    # Load all lines from file
    list_all = [x for k,x in enumerate(f.readlines())]
        
# Create/Overwrite and start writing log file (A new logfile will be created each day
# the script is run)   
with open(path2results + log_filename, 'w') as log_f:
    log_f.write('Import date: ' + today.strftime("%d-%m-%y") + '\n')
    log_f.write(list_all[0])

# Open to append logs
log_f = open(path2results + log_filename, 'a') 

# Locate trial line boundaries and first trial line
list_tr_bb = [n for n, x in enumerate(list_all) if re.match(r'^Summary Trial', x)]
trial_headers = [x for n, x in enumerate(list_all) if re.match(r'^Summary Trial', x)]

log_f.write( str(len(list_tr_bb)) + ' trials found.\n' )

# Add a final entry to list_tr_bb to mark end boundary of last trial
list_tr_bb.append(len(list_all))
```

```python
### DATA CLENANING
# Data files contain duplicate lines and/or duplicate trials
# Find and remove them

# Look for duplicate trials to write info to log
dupe_trials = [x for n, x in enumerate(trial_headers) if x in trial_headers[:n]] # Get duplicate trial headers

dupe_trls_mess = 'Found ' + str(len(dupe_trials)) + ' duplicate trials: \n'

print(dupe_trls_mess)
[print(x.split(' ')[1] + ' ' + x[-3:-1]) for x in dupe_trials]
log_f.write(dupe_trls_mess)
[log_f.write(x) for x in dupe_trials]

# Look for duplicate lines in general and remove them from list_all
dupe_lines = [n for n, x in enumerate(list_all) if x in list_all[:n]]

# Remove duplicate lines from list_all
dupe_ln_mess = 'Found ' + str(len(dupe_lines)) + ' duplicate lines.\n'
print(dupe_ln_mess)
log_f.write(dupe_ln_mess)
for x in sorted(dupe_lines, reverse=True):
    # log_f.write('line: ' + str(x) + '\n')
    del list_all[x]

# Update trial boundaries after deletions
list_tr_bb = [n for n, x in enumerate(list_all) if re.match(r'^Summary Trial', x)]

log_f.write('Number of trials left after deletions: ' + str(len(list_tr_bb)) + '\n')

# Add a final entry to list_tr_bb to mark end boundary of last trial
list_tr_bb.append(len(list_all))
```

```python
## SEPARATE VARIABLES, and save data as a series of three nested lists: 
# 1. List of trials e.g. list_sess[0]: All rows from trial 1
# 2. List of rows representing movemementfrom one node to the next. 
#    e.g. list_sess[0][0]: First line from first trial: first movement
# 3. List of data within each row. e.g. list_sess[0][0][0]: First trial, 
#    First row (movement), Starting node

# Initialize list of lists, and remove extra characters (such as brackets and line breaks),
# leaving only spaces as var separators

list_sess = []
for l,tr in enumerate(list_tr_bb[:-1]): 
    # Find all lines starting with ( between trial boundaries
    list_sing = [x for x in list_all[list_tr_bb[l]:list_tr_bb[l+1]] if x[0]=='('] # Lines starting with ( 
    # Remove parenthesis, commas and '' from string
    list_sing = [x.replace("(", "") for x in list_sing]
    list_sing = [x.replace(")", "") for x in list_sing]
    list_sing = [x.replace(",", "") for x in list_sing]
    list_sing = [x.replace("'", "") for x in list_sing]
    list_sing = [x.replace("\n", "") for x in list_sing]
    # Now separate data in each line within the trial
    list_sing = [x.split(sep=' ') for x in list_sing]
    list_sess.append(list_sing)
    
final_message = 'Number of trials loaded: ' + str(len(list_sess)) + '\n'
print(final_message)
log_f.write(final_message)

log_f.close()
```

```python
## TRANSFORM TO 'LONG' FORMAT AND SAVE IN PANDAS DATA FRAME
# Transform current 'wide' data format into 'long' format, and store as a pandas 
# dataframe.
# Edge trajectory data will be re-ordered into a successive list of nodes and times.
# Additional data, such as seconds, distance, and speed refers to the trajectory from
# node[i-1] to node[i], being zero for the inital row.

# First re-order and store as single lists for each variable
size = sum([len(x) for x in list_sess]) + len(list_sess) # array shape: total no. of lines (across all trials), 
                                                # with one additional line per trial (25 trials in total) to flatten structure
trial_no = [None]*size #np.zeros(size, dtype=int)
distance = [None]*size #np.zeros(size) # cumulative distance travelled within each trial
seconds_ff = [None]*size #np.zeros(size) # cumulative time within each trial
nodes = [None]*size #np.zeros(size, dtype=int) # nodes the rat passes through (flat)
speed_ff = [None]*size #np.zeros(size)
times = [None]*size #np.zeros(shape) # all timestamps (flat)


line_no = 0 # initialize counter

for tr, trial in enumerate(list_sess):

    # Add inital trial line for nodes and timestamps array with starting node (first value of first line in the trial)
    trial_no[line_no] = tr+1
    nodes[line_no] = trial[0][0] # First value from first row in that trial: start node
    times[line_no] = trial[0][2] # Third value from first row in that trial: start time
    # times[:, cum_line_no] = [float(x) for x in time_list]
    # Initialize distance and seconds
    distance[line_no] = 0
    seconds_ff[line_no] = 0
    speed_ff[line_no] = 0
    line_no +=1
    
    for row in trial:
        
        trial_no[line_no] = tr+1
        nodes[line_no] = row[1]
        times[line_no] = row[3]
        seconds_ff[line_no] = row[4]
        distance[line_no] = row[5]
        speed_ff[line_no] = row[6]     
        line_no += 1


## BUILD PD DATA FRAME and ADD USEFUL COLUMNS

data = pd.DataFrame(list(zip(trial_no, nodes, times, distance, seconds_ff, speed_ff)),
                   columns = ['trial_no', 'nodes', 'times_ff', 'distance', \
                              'seconds_ff', 'speed_ff']) 

# Set trial number as index for easy selecting trials
data.set_index('trial_no', inplace=True)

# Set appropriate data types for each column. Times data set to timedelta object,
# although maybe it would better be just a time? (day data is not relevant in our case)
data["distance"] = data["distance"].astype('float')
data["seconds_ff"] = data["seconds_ff"].astype('float')
data["speed_ff"] = data["speed_ff"].astype('float')
data["times"] = pd.to_timedelta(data["times_ff"])


# Add column with cumulative distance and cumulative seconds
data["cum_distance"] = ""
data["cum_seconds"] = ""
for tr in range(1, len(list_sess)):
    data.loc[tr, "cum_distance"] = data.loc[tr, "distance"].cumsum()
    data.loc[tr, "cum_seconds"] = data.loc[tr, "seconds_ff"].cumsum()

# Add column with node classification    
data["island"] = ""
data.loc[data["nodes"].astype('int') // 100 == 1, "island"] = 'Ireland'
data.loc[data["nodes"].astype('int') // 100 == 2, "island"] = 'Japan'
data.loc[data["nodes"].astype('int') // 100 == 3, "island"] = 'Hawaii'
data.loc[data["nodes"].astype('int') // 100 == 4, "island"] = 'Easter I.'

data.info()
data.head()
       
```

### Perform basic sanity checks and remove bad trials

* All times must be consecutive and descending
* The last node on every trial must be 104
* Node 104 *must not appear* in the middle of a trial

We find there are three bad trials, where node 104 is found in the middle of the trial. These are the first, the last, and the middle trial (1, 14 and 28). Interestingly, this was also the case in the other datafile, with bad trials being 1, 11 and 25 (also first, last, and one middle one, although not *the* middle one) 

```python
## SANITY CHECKS

# Check the times are all consecutive and descending
if ~(data['times'].sort_values() == data['times']).all():
    print('ERROR: Times are not monotonically descending')

bad_trials = []
# Check that 104 is the last node of every trial, and doesn't appear anywhere else
for tr in range(1, data.index.max()+1):
    these_nodes = data.loc[tr, 'nodes']
    if these_nodes.iloc[-1] != '104':
        print('ERROR: trial ' + str(tr) + 'does not end in node 104')
        bad_trials.append(tr)
    if (these_nodes.iloc[:-2] == '104').any():
        print('ERROR: trial ' + str(tr) + ' contains target node 104 in places other than last')
        bad_trials.append(tr)

# Remove bad trials from dataframe
data_cl = data.drop(bad_trials)
print('\nRemoved bad trials: ' + str(bad_trials) + '\n')
data_cl.info()
```

### U-turns and deviation from optimal (shortest) path

We will now look for U-turns and examine deviation from optimal (shortest path).

We find there are 6 u-turns, one in each trial: trial_nos: 2, 5, 10, 13, 16 and 24. U-turns always occur in starting island: In the bar chart, bar colour indicates starting island. In turn, each bar is annotated with the U-turn node and the island it belongs to. Both things co-incide.

We realise re-visiting a node could be an interesting moment, even if it is not an immediate u-turn. We look for duplicate nodes in the trajectory. Trials 2 and 4, which are also the longest, have several re-visits.

```python
# Detect u_turns: moments when current node is the same as 2 nodes ago.
# Store as a boolean column to the main datafrmae
data_cl['u_turn'] = (data_cl['nodes'] == data_cl['nodes'].shift(2))

# Select steps that are u-turns
u_turns = data_cl[data_cl['u_turn'] == True]

print('U-turns detected: ')
u_turns

# Plot number of u-turns per trial.
# I need to reset index, so that trial_no is a standard column and I can plot it
u_turns = data_cl[data_cl['u_turn'] == True].reset_index()

sns.countplot(x='trial_no', data=u_turns,\
             palette = starting_island[u_turns['trial_no'].unique()].map(color_map).values)
plt.title('Trials with U-turns colour-coded according to STARTING island')
plt.ylabel('Number of U-turns annotated with U-turn node')
plt.legend(handles[:-1], labels[:-1])

# Annotate longest trials
tr_label_1 = u_turns['island'].to_list()
tr_label_2 = u_turns['nodes'].to_list()
for i, txt1 in enumerate(tr_label_1):
    txt = txt1 + ':\n' + tr_label_2[i]
    plt.annotate(txt, (i-0.3, 0.5))
```

```python
## Find nodes visited twice in each trial

data_cl['re-visited'] = False

for tr in data_cl.index.unique():
    dup_nodes = data_cl.loc[tr, 'nodes'].duplicated()
    data_cl.loc[tr, 're-visited'] = dup_nodes
    
data_cl[data_cl['re-visited'] == True] \
            [['nodes', 'speed_ff']]

# Plot number of re-visits per trial.
# I need to reset index, so that trial_no is a standard column and I can plot it
re_visited = data_cl[data_cl['re-visited'] == True].reset_index()

sns.countplot(x='trial_no', data=re_visited, \
             palette = starting_island[re_visited['trial_no'].unique()].map(color_map).values)
plt.title('Trials with re-visited nodes colour-coded according to STARTING island')
plt.ylabel('Number of node re-visits')
plt.legend(handles[:-1], labels[:-1])
```

```python
# Calculate distance to optimal path at each step (distance measured in steps)
# Add column with optimal and actual distance at each step

data_cl['act_stps_2trgt'] = np.nan
data_cl['min_stps_2trgt'] = np.nan

#line_no = 0
for tr in data_cl.index.unique():
    # Number of steps taken by rat in the trial (trajectory lenght):
    traj_len = len(data_cl.loc[tr, 'nodes'])
    
    # Number of steps still left until rat reaches target, at each node passed:
    act_stps_2trgt = [traj_len-n for n in range(1, traj_len+1)]
    
    # Minimum number of steps until target at each node passed by rat
    min_stps_2trgt = [DD[str(a_node)]['104'] for a_node in data_cl.loc[tr, 'nodes']]
                
    data_cl.loc[tr, 'act_stps_2trgt'] = act_stps_2trgt
    data_cl.loc[tr, 'min_stps_2trgt'] = min_stps_2trgt
    #dist_2_opt[trial_l == tr] = np.array(tr_trajectory) - np.array(tr_opt_traj)
    
    
data_cl['act_stps_2trgt'] = data_cl['act_stps_2trgt'].astype('int')
data_cl['min_stps_2trgt'] = data_cl['min_stps_2trgt'].astype('int')
    
```

## Shortest distance plots

Some observations:

- In ***Trial 2***, rat seems pretty lost in Japan, although starting island is Easter. I.
- At bridges, rat seems to always speed-up

```python
# Plot u_turns and re-visited nodes in shortest distance plots

# Generate list of labels to annotate plots
lbls = []
for i, node in enumerate(data_cl['nodes'].astype('str')):
    if data_cl.iloc[i]['u_turn']== True:
        lbls.append('U_TURN: ' + node)
    elif data_cl.iloc[i]['re-visited'] == True:
        lbls.append('RE: ' + node)
    else:
        lbls.append("")

# Distribute plots in two columns
n_cols = 2
n_rows = 2 #int(len(data_cl.index.unique())/n_cols +0.5) #(add 0.5 to round upwards)
n_tr = 0
tr = 0

while tr < data_cl.index.unique().max():
    fig, ax = plt.subplots(n_rows, n_cols)
    plt.legend(handles, labels)
   
    for row in range(n_rows):
        for col in range(n_cols):
            if n_tr == len(data_cl.index.unique()):
                break
            tr = data_cl.index.unique()[n_tr]
            x = list(range(0, len(data_cl.loc[tr])))
            y = data_cl.loc[tr, 'min_stps_2trgt']
            #ax[row, col].plot(x, y, marker = 'o', label=str(tr));
            sns.lineplot(ax=ax[row, col], x=list(range(len(data_cl.loc[tr]))), \
                            y = 'min_stps_2trgt', hue='island', marker = 'o',\
                            data = data_cl.loc[tr], palette = color_map_rgb, label=str(tr), legend=False)
            
            trl_idx = [i for i, is_true in enumerate(data_cl.index == tr) if is_true]
            for j, i in enumerate(trl_idx):
                ax[row, col].annotate(lbls[i], (x[j], y.iloc[j]))

            ax[row, col].set_title('Trial ' + str(tr))
            #ax[row, col].legend();

            ax2=ax[row, col].twinx()
            # make a plot with different y-axis using second axis object
            ax2.plot(x, data_cl.loc[tr, 'speed_ff'], \
                           color="gray",marker="x")
            

            # Add hypothetical speed phase boundaries
            ax2.plot([0, max(x)], [0.25, 0.25], 'k:')
            ax2.plot([0, max(x)], [0.5, 0.5], 'k:')

            n_tr += 1
            
            
        ax2.set_ylabel("Speed", color="gray", fontsize=14)
        ax[row, col-1].set_ylabel('Shortest distance', color="blue",fontsize=14);
        
    ax[1, 0].set_xlabel('Step');
    ax[1, 1].set_xlabel('Step');
    


            #plt.show()
    
    
   
    #plt.pause(0.3)        
        
```

```python
# Now plot deviation from shortest path for each trial as a bar graph, and add
# deviation value as a new column in dataframe

data_cl['deviation'] = data_cl['act_stps_2trgt'] - data_cl['min_stps_2trgt']

x = data_cl.index.unique()
y = data_cl.groupby('trial_no')['deviation'].first()

plt.bar(x, y, color = starting_island.map(color_map))
plt.xticks(x)
plt.legend(handles[:-1], labels[:-1])
plt.title("Deviation from optimal path coloured according to starting island")
# I love that there seem to be cyclic oscillations in performance!!
```
